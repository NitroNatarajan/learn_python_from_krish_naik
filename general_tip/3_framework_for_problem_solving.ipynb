### Framework for solving the problem 
  - Process to solve a problem 
    1. Analyze the problem
       1. Nature of the input, output and constraint like size or range of values.
    2. Break down the problem into smaller sub-parts 
    3. Remember/enlist the concept like list or set, dict.
    4. Take 2-3 example to make sure understand the problem and clear out the confusion
    5. Write a pseudocode before the actual code to be written
    6. DRY run it once.
    7. Write down the solution.
    8. Look out for the edge cases/boundary condition.
