#### Map Function
  - The map() function applies a given function to all the items in an input list(or any other iterables) and return the map object(an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
# Map function 
# Syntax: map(function(), input_list_or_iterables)
def square(x):
  return x*x
print(square(3))

# In case if we have a list of numbers, we wanna double the numbers.
# We can apply this function to all the numbers in an array.
numbers = [4,5,6,7,8,9]
double = list(map(square, numbers))
print(double)
```

```python
# Lambda function
numbers = [1,2,3,4,5,6,7]
squared = list(map(lambda x:x*x , numbers))
print(squared)
```

```python
# using map function with multiple iterables 
num1 = [1,2,3,4,5,6]
num2 = [2,3,4,5,6,7]

sum_list = list(map(lambda x,y: x+y , num1 ,num2))
print(sum_list)
```

```python
# Map to covert the list number strings to integer 
str_numbers = ['1','2','3','4','5']
integer = list(map(int,str_numbers))
print(integer)
```

```python
words = ['apple','banana','cherry']
upper_words = list(map(lambda x:x.upper() , words))
upper_words1 = list(map(str.upper , words))
print(upper_words)
print(upper_words1)
```

```python
# Its a way to return the name from the person dictionary 
def get_name(person):
  return person['name']

# For eg: we have list of dictionary in a list, we must need every persons name.
persons = [
  {"name":'Natraj','age':32, 'location':'vilagam'},
  {"name":'Kari','age':32, 'location':'vilagam'},
  {"name":'Thanaraj','age':32, 'location':'vilagam'},
  {"name":'Leelavathi','age':32, 'location':'vilagam'}
  ]

names_list = list(map(get_name, persons))
print(names_list)
```
