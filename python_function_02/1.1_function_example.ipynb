```python
# Temperature conversion 
# logic: 
#  f = (c*(9/5) + 32)

def temp_conversion(temp, unit):
  if unit.lower() == 'c':
    x = temp* (9/5) + 32
    return x
  elif unit.lower()== 'f':
    y = (temp-32)*(5/9)
    return y

f_c = temp_conversion(120,'f')
c_f = temp_conversion(30,'c')

print(f_c , c_f)
```

```python
# Password Strength checker 

def password_strength_check (password):
  if len(password) < 8 :
    return False
  if not any(i.lower() for i in password):
    return False
  if not any(i.upper() for i in password):
    return False
  if not any(i.isdigit() for i in password):
    return False
  if not any(i in '!@#$%^&*_+. ' for i in password):
    return False
  return True

print(password_strength_check('Natarajan0501'))
print(password_strength_check('Natarajan0501.'))
print(password_strength_check('Nitro1.   '))
```

```python
# Calculate the total cost of an item in the cart 
def total_cost_cart(cart):
  total_cost = 0 
  for item in cart:
    total_cost += item['price']*item['quantity']
  return total_cost
    

# cart
cart = [{'name': 'apple', 'price':1, 'quantity':2},
        {'name': 'orange', 'price':2, 'quantity':1},
        {'name': 'guava', 'price':3, 'quantity':4}]

total = total_cost_cart(cart)
print(total)
```

```python
# Check if a string is palindrome or not:

def is_palindrome(s):
  s=s.lower().replace(" ","")
  return s==s[::-1]

print(is_palindrome('Masam'))
```

```python
# Factorial 
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)
  
print(factorial(7))
```

```python
# A function to read the file and count the frequency of each word.

def count_word_frequency(file_path):
  word_count={}
  with open(file_path) as file:
    for line in file:
      words = line.split()
      for word in words:
        word=word.lower().strip('.,!?\'"')
        word_count[word]=word_count.get(str(word),0)+1
  return word_count

file_path = 'nitro.txt'
print(count_word_frequency(file_path))
```

```python
# Validate the email address 
import re
def is_valid_email(email):
  pattern = r'^[a-zA-Z0-9_.+-]+@[A-Za-z0-9-]+\.[a-zA-Z0-9]+$'
  return re.match(pattern,email) is not None

# Calling a function 
print(is_valid_email('natarajan0501@gmail.com'))
print(is_valid_email('natraj.com'))
```
