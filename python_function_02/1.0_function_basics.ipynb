#### Introduction to python function:
  - Function is kind of task doer, is wrapper of some code which work with the input and kind of generate the output based on the logic and it return the output.
  - Generally, we use this way to reuse the code. Once we defined the function, we can reuse any no of times and by calling the function just by pass the values in the input, we can get the output as the logic we defined inside the function definition.
  - In short, function is a block of code that perform a specific task. It helps in organizing the code, reusing the code and improve the code readability.
  - First we need to define the function with `def` python keyword, here is where we write the logic.
  - By passing the arguments while calling the function after the function definition, we can able to get the result. 

```python
# Function definition
# Syntax function 
def function_name(parameter):
  '''code block with logic'''
  # Function body where the code logic will be written
  return parameter;

print(function_name, type(function_name))
```

```python
# Why function?
# if suppose i want to check whether the given number is odd or even:

# Function definition
def even_or_odd(num):
  if num%2 == 0 : 
    return f'The given num {num} is even'
  else:
    return f'The given num {num} is odd'

# Function calling with the input 
print(even_or_odd(20)) # it return the result 
print(even_or_odd(11)) 
```

```python
# Function with multiple parameter 
def add(a,b):
  return a + b;
result = add(2,3)
print(result)
```

```python
# Default Parameter 
def greet(name='Dude'):
  return f'Hey {name}.. Welcome to the heaven'

print(greet('Natarajan'))
print(greet())
```
