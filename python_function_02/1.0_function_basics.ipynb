#### Introduction to python function:
  - Function is kind of task doer, is wrapper of some code which work with the input and kind of generate the output based on the logic and it return the output.
  - Generally, we use this way to reuse the code. Once we defined the function, we can reuse any no of times and by calling the function just by pass the values in the input, we can get the output as the logic we defined inside the function definition.
  - In short, function is a block of code that perform a specific task. It helps in organizing the code, reusing the code and improve the code readability.
  - First we need to define the function with `def` python keyword, here is where we write the logic.
  - By passing the arguments while calling the function after the function definition, we can able to get the result. 

```python
# Function definition
# Syntax function 
def function_name(parameter):
  '''code block with logic'''
  # Function body where the code logic will be written
  return parameter;

print(function_name, type(function_name))
```

```python
# Why function?
# if suppose i want to check whether the given number is odd or even:

# Function definition
def even_or_odd(num):
  if num%2 == 0 : 
    return f'The given num {num} is even'
  else:
    return f'The given num {num} is odd'

# Function calling with the input 
print(even_or_odd(20)) # it return the result 
print(even_or_odd(11)) 
```

```python
# Function with multiple parameter 
def add(a,b):
  return a + b;
result = add(2,3)
print(result)
```

```python
# Default Parameter 
def greet(name='Dude'):
  return f'Hey {name}.. Welcome to the heaven'

print(greet('Natarajan'))
print(greet())
```

```python
# Variable length arguments

# 1. Positional argument 
# When we have multiple arguments, we can use the positional argument to reference those arguments.

# For eg:
def print_numbers(*args):
  print(args)
  for number in args:
    print(number)

# I am gonna call the function with args 
print_numbers(1,2,3,4,5,6,'Natraj',2.8765)
# Here the args keyword reference the arguments.
# Here we can use any word instead of args.
```

```python
# Keyword argument 
# If the args are in the form of key value pairs, the n instead of positional args we can use the kwargs. 

# eg:
def print_details(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")

print_details(name='Natraj',age=31,grade='A')
```

```python
# Lets combine the positional and keyword args 

def print_args(*args,**kwargs):
  for i in args:
    print('Positional args : ', i)
  for key,value in kwargs.items():
    print(f"{key}:{value}")

print_args(1,2,3,4,5, name='Natraj', age=32,grade='B')

```

```python
# Return Statement
def multiply(a,b):
  return a*b

mu1 = multiply(2,3)
print(mu1)

# Return multiple values from the single return statement
def multiply_remainder(a,b):
  return a*b,a%b

mul,rem = multiply_remainder(2,3)
print(mul)
print(rem)
```
