### String Methods 

```python
# str datatype definition and getting help for the methods 
dir(str) # it list out all the available methods inside the str datatype object 
help(str.lower) # it illustrate the definition of the methods 
```

```python
# Type checking after the string object creation with the str construction function 
string = str('nitro')
type(string)

num_string = str(22)
deci = str(23.9)

print(num_string)
print(type(num_string))
print(deci)
print(type(deci))
```

```python
# _add_ 
str.__add__('a','b') # 'ab'

# __contains__
str.__contains__('nitro','ni') # true

# __eq__
str.__eq__('a','b') # false 

# __ge__
str.__ge__('02' , '03') # true 

# __le__
str.__le__('32', '33')

# __ne__
str.__ne__('21','12')

# __gt__
str.__gt__('32', '12')

# __hash__
str.__hash__('23')

# __len__
str.__len__('nitronatraj')

# __lt__ 
str.__lt__('32', '28') 

# __mod__
# str.__mod__('15','14')

# __mul__
str.__mul__('12',11)


# __sizeof__
str.__sizeof__('natraj')

# __str__
str.__str__('nat')
```

```python
# Capitalize 
string = 'HI I AM NATARAJAN. \n I am karikalan '
string.capitalize()
```

```python
# casefold 
string1 = "Hello World"
string2 = "hello world"
if string1.casefold() == string2.casefold():
    print("Strings are equal (case-insensitive)")
```

```python
# center, ljust, rjust
name = 'Natraj'
name.center(30, '*')
name.ljust(20, '-')
name.rjust(30,'%')
```

```python
# count 
name = 'natarajan'
name.count('n')
name.count('a')
name.count('a',4)
```

```python
# encode
name='nitro'
name.encode('utf-8')
name.encode('utf-16')
name.encode('utf-32')
```

```python
# endswith 
name = 'hello'
name.endswith('llo')
name.endswith('llol')
```

```python
# expandtabs
text_with_tabs = "Name:\tAlice\nAge:\t30\nCity:\tNew York"

# Using default tabsize (8)
expanded_text_default = text_with_tabs.expandtabs()
print("Default tabsize (8):\n", expanded_text_default)

# Using a custom tabsize (4)
expanded_text_custom = text_with_tabs.expandtabs(tabsize=4)
print("\nCustom tabsize (4):\n", expanded_text_custom)
```
