Introduction to Conditional Statements: 
  - Generally, we write a logic to execute if some of the conditions were met. In that case we use this conditional statements. In this statements, we can construct the logic accordingly what need to be happen if the condition met or not.
  - 

```python
#  Conditional statement 
print("======== if block ============")
age = 20 
if age >= 18:
  print(f"You are allowed to vote in the election. ")


# else statement: if the if block condition is not true, then the else block will be executed. 
print("============ if-else ==============")

age = 17 
if age >= 18:
  print(f"You are allowed to vote in the upcoming election. ")
else:
  print(f'You have to wait about {18 - age} years to get eligible to vote..')

print("============ if-elif-else ==============")
# if-elif-else block 

# if we have three group of condition to be tested for three different actions to be performed, then we can go with 
# if-elif-else statement. 


team_a = 'blue'
team_b ='yellow'
team_c = 'red'
winner = 'yellow'

if winner == team_a:
  print(f'{team_a} house is the winner.')
elif winner == team_b:
  print(f"{team_b} house is the winner.")
else:
  print(f"{team_c} house is the winner.")
```

```python
# Nested if statements 
# It means using the if-elif-else inside the if-elif-else statement. 
# We can able to nest the conditionals to any level deeper as we need. 

# number even or odd or negative

num = int(input("Enter the number?"))

if num > 0 :
  print(f'The number {num} which is positiveeee')
  if(num % 2 == 0):
    print(f"The given number {num} is even")
  else:
    print(f"The given number {num} is odd")
elif num < 0:
  print(f'The number {num} which is negative')
  if(num % 2 == 0):
    print(f"The given number {num} is even")
  else:
    print(f"The given number {num} is odd")
else:
  print(f"The given number {num} is even and also neither positive nor negative")

```

```python
year = int(input("Enter the year?"))

if (year % 4 == 0 and year%100 != 0) or (year%400 == 0) :
  print(f"The year {year} is leap year..")
else:
  print(f"The year {year} is not a leap year.")
```

```python
# Simple calculator 
num1 = int(input('Enter thee first number: '))
num2 = int(input('Enter thee second number: '))
operator = input("Enter the symbol of any one of the operation in this list ->  +, - , / , * , ** , // ")
result = 0; 
if operator == '+' :
  result = num1 + num2;
  print(result)
elif operator == '-':
  result = num1 - num2
  print(result)
elif operator == '*':
  result = num1 * num2 
  print(result)
elif operator == '**':
  result = num1 ** num2
  print(result)
elif operator == '/':
  if num2 != 0 :
    result = num1/num2
    print(result)
  else:
    print("Denominator should not be zero")
elif operator == '//':
  if num2 != 0 :
    result = num1//num2
    print(result)
  else:
    print("Denominator should not be zero")
else:
  print('Invalid Operator')
```
