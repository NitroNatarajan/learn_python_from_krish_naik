### Data type in python:
  - Since the data is classified differently based on its nature, the data type which tells the compiler or the interpreter about how the programmer intend to use the data.
  - By defining the data type of the variable, interpreter determine the type of operations performed on the data, what kind of possible values it could take or allowed to assign and also which help the interpreter to dynamically allocated the space in the memory location based on the data type.

### Importance of the data types in programming: 
   - It ensures that the efficient memory allocation.
   - And it allows the compatible values to be assigned and possible operations on that data.
   - Proper use of the data type will prevent the code from bugs and errors. 

### Data types: 
  1. Basic Data types in python 
     1. String (str)
     2. Integer (int)
     3. Float (float)
     4. Boolean (bool)
  2. Advanced Data TYpes 
     1. List [ ]
     2. Tuple ( )
     3. Dictionary { }
     4. Sets  

```python
# Integers datatype 
print("=== Integer ===")
age = int(23)
type(age)
print(age , type(age))

print("=== Float Data type ===")

width = float(25.4)
print(width, type(width))
```

```python
print("=== String Data Type ===")
first_name = 'Nitro'
last_name = str('Natarajan')
print(first_name, last_name)
```

```python
print("First_name: " , first_name , type(first_name))
print("Last_name: " , last_name , type(last_name))
```

```python
print("=== Boolean Data Type ===")
is_he_graduate = True
is_he_adult = bool('Yes')
print("is_he_graduate: " , is_he_graduate , type(is_he_graduate))
print("is_he_adult: " , is_he_adult , type(is_he_adult))

# Each constructor function hold some properties and methods into it. We will see it in depth as we go further.


print('Default return value of the bool constructor is: ' , bool())
```

```python
# Common Error:
# concatenating error 

# result  = age + first_name 
# print(result)
# It return the error because there is conflict between data types of operands while concatenating.

result = str(age) + first_name

print(result)
```
