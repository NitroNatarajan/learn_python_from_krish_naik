# Introduction to the dictionaries 
  - Dictionaries are unordered collection of items. They store data in key-value pairs. 
  - Keys must be unique and immutable 
  - Values could be anytype like string, tuple, dictionary itself and lst or set.. 

```python
# Creating the dictionaries 
empty_dict = {}
empty_dict1 = dict()
new_dict = {"name": "Natarajan",'age': 23}

print(empty_dict , type(empty_dict))
print(empty_dict1, type(empty_dict1))
print(new_dict, type(new_dict))

student_details = {"Name": "Natarajan",
"age": 31, "Grade": 'A' , "age": 25}
# If we use the duplicate key name, then the value will be replace the first in key's value
print(student_details)
```

```python
# Single key will be used always 
student = {'a': 1, 
           'b':2 , 'a': 4}
print(student)

empty_dict = {}
print(empty_dict, type(empty_dict))
empty_dict = dict()
print(empty_dict, type(empty_dict))

empty_set = set()
print(empty_set, type(empty_set))

```

```python
# Accessing the dictionary elements 

student_details = {"name": "Natarajan",
"age": 31, "Grade": 'A' }

name = student_details['name']
print('Name :', name)
age = student_details['age']
print('Age:', age)

# We can able to access the elements of dictionary by get method 
name = student_details.get('name')
age = student_details.get('age')
print(name, age)

# What if i try to get the value of undefined key name in the dictionary 
last_name = student_details.get("last_name")
print(last_name) # It gives none. I can able to return the different value if no key name is exist as i asked. 
last_name = student_details.get('last_name','Not available')
print(last_name)

# We can able to access the elements in the dictionary by direct calling or by get method.
```

```python
# Modifying the dict elements 
# Dictionary is mutable which means we can able to update, delete or add elements. 
# But the duplicate key is not allowed, if we try with the duplicate key, the value will get replaced.
print(student_details)
student_details['name'] = 'Nitro Natarajan' # we can update the name and age 
student_details['age'] =32
print(student_details)

# Lets add some new key value pair into the student_details 
student_details['address'] = 'Tamilnadu India'
print(student_details)

# Lets delete the elements in the dictionary 
print(student_details)
del student_details['age']
print(student_details)

# So we can able to update the values, delete the value add the new key value 
```

```python
# Dictionary methods 

keys = student_details.keys() # keys 
values = student_details.values() # values
print(keys)
print(values)
items = student_details.items() # all the items in the form of list of tuples
print(items)
```

```python
# copy
student_details_copy = student_details 

print(student_details)
print(student_details_copy)
student_details_copy['Grade'] = 'B'
student_details['name'] = 'Nitro'
print(student_details)
print(student_details_copy)

# When we use this way of copy, both the dict mutated when we update any single dict 
```

```python
# So we have a way to fix this thats called the shallow copy.
print(student_details)
student_copy = student_details.copy()
student_copy['name'] = 'Karikalan'
print(student_details)
print(student_copy)

# Shallow copy provide us the different memory location for the values get referenced from, 
# so that the reference object never get spoiled.
```

```python
# Iterating over the dictionaries 
# We can use the loops to iterate over the keys, values or items as well.

print(student_copy)
for key,value in student_copy.items():
  print(key,value)

for key in student_copy.keys():
  print(key)

for value in student_copy.values():
  print(value)
```

```python
# Nested dictionaries -> means dict inside dict 
students = {
  'student1':{
    'name':'Natraj',
    'age':31,
    'grade':'A' 
     },
  'student2':{
    'name':'Karikalan',
    'age':29,
    'grade':'B' 
   }
}

# How to access the nested dictionary elements 
print(students['student2']['name'])
print(students['student1']['grade'])
```

```python
# Iterating over the nested dictionary
students = {
  'student1':{
    'name':'Natraj',
    'age':31,
    'grade':'A' 
     },
  'student2':{
    'name':'Karikalan',
    'age':29,
    'grade':'B' 
   }
}

for student_id , student_info in students.items():
  print(student_id)
  for key,value in student_info.items():
    print(key,value)
  print("==========")
```

```python
# Dictionary comprehension 
squares = {x:x**2 for x in range(5)}
print(squares)

# Conditional dict comprehension 
evens = {x:x for x in range(5) if x%2==0}
print(evens)
```

```python
# Practical examples 
numbers = [1,2,2,3,4,4,4,4,4,5,5,6,6,6,7,7,7,8,9,9,9]
# Using the dictionary to count the frequency 
frequency = {}
for i in numbers:
  if i in frequency:
    frequency[i] += 1 
  else:
    frequency[i] = 1

print(frequency)
```

```python
# Merge two dict into one 
dict1 = {'a':1, "b":2}
dict2 = {'c':3, 'd':4}
merge_dict = {**dict1, **dict2}
print(merge_dict)
```
