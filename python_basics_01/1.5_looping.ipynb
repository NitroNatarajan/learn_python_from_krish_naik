```python
# Looping statement is used to execute a block of code repeatedly until a certain condition is met.

# There are two types of loops in Python: for loop and while loop.
# 1. For Loop: It is used to iterate over a sequence (like a list, tuple, dictionary, set, or string).

# 2. While Loop: It is used to execute a block of code as long as a specified condition is true. 

# Here are examples of both types of loops:
# Example of For Loop
name = "Alice"
for char in name:
    print(char)

# Example of While Loop
count = 0
while count < 5:
    print("Count is:", count)
    count += 1

# Example of For Loop with range()
for i in range(1, 6):
    print("Number:", i) 

# Example of While Loop with break statement
num = 1
while True:
    print("Number is:", num)
    if num >= 5:
        break
    num += 1

# Example of For Loop with continue statement
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print("Odd Number:", i) 

# Example of While Loop with continue statement
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue
    print("Odd Number:", n)

# Note: Be cautious with while loops to avoid infinite loops by ensuring the condition will eventually become false.

# Example for the pass statement in loops
for i in range(5):
    pass  # Placeholder for future code
# The pass statement is used when a statement is required syntactically but no action is needed.
# It allows you to create minimal classes or functions without any implementation.
# It is also useful in loops when you want to create a loop structure but do not want to execute any code inside it yet.

# Example of pass statement in a while loop
i = 0
while i < 5:
    pass  # Placeholder for future code
    i += 1
# The pass statement does nothing and is just a placeholder.
```

```python
# Nested loops:
# A loop inside a loop 

for i in range(1,11):
  print(f'Table Number:{i} starts')
  for j in range(1,11):
    print(f'{j}x{i}={j*i}')
  print(f'Table number: {i} ends')
  print("==========================")
```

```python
# Calculate the sum of first n natural numbers 

# for loop 
number = int(input("Enter the natural number to get its first n natural number?"))
count = 0;
for i in range(1,number+1):
  count += i
print(count)

count=1
sum = 0
while count <= number:
  sum += count
  count += 1
print(sum)
```

```python
# Prime number 
# Print the prime numbers between 0 to specified number:
number = 100
for i in range(1,number+1):
  if i == 1 :
    print(1)
  else:
    for j in range(2,i):
      if i%j==0:
        break
    else:
      print(i)
```

```python
# prime number logic using the looping statement 
'''
The for-else statement in Python provides an optional else block that is executed only if the for loop completes 
all its iterations without encountering a break statement. This structure is useful for scenarios where a specific
action needs to be performed if a certain condition is not met during the loop's execution.
'''
# prime number from 1 to 100 
for i in range(1,101):
  if i == 1:
    print(i)
  else:
    for j in range(2,i):
      if i%j == 0 :
        break
    else:
      print(i)
```

```python
'''
In case if want to run the loop for a finite no of times without using the iteration value count,
In that case i can use the _ in the iterator counter.
'''
# Like 
# eg: If i want to repeat write the same statement thrice

for _ in range(3):
  print('Natarajan') 

```
