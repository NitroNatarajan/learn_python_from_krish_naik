Operator in Python:
  Generally, we use the operators to build the logics for the specific task to happen. 
  - Introduction to operators 
  - Arithmetic operators
  - Comparision operators
  - Logical Operators

```python
# Arithmetic operations using the arithmetic operator 

a = 2
b , c = 5 ,3 ; 
print(a,b,c)

# Addition 
add = a +b ;
print('add: ', add)

# Subtract 
sub = a -b 
print('sub : ', sub)

# Multiply 
multiply = a * b ;
print("multiply : " , multiply)

# division:
div = a/b  
print('division: ', div)

# floor division
f_div = a // b
print('floor_division: ' , f_div)


# Modulus -> means the remainder during division
mod = a%b
print('Modulus : ' , mod)

# Exponential Operation: 
exp = a ** 2 
print("Exponential Operation Result : ", exp)
```

```python
# Comparision Operator 

#  == (Double Equal to)
a = 10 
b = 10 

equal_to = a == b;
print("equal_to : " , equal_to)

name = "Nitro"
Name = "Nitro"
is_string_equal = name == Name
print("is_string_equal :" , is_string_equal)

# != -> Not equal to 
is_not_equal = a != b 
print('is_not_equal: ', is_not_equal)

is_unmatched_strings = name != Name
print("is_unmatched_strings :" , is_unmatched_strings) 


# Greater than 
a = 20
b = 30
is_greater = a > b
print('is_greater:', is_greater)


# Greater than or equal to  
a = 20
b = 20
is_greater_than_or_equal = a >= b
print('is_greater_than_or_equal:', is_greater_than_or_equal)


# Lesser than 
a = 20
b = 30
is_lesser = a < b
print('is_lesser :', is_lesser)

# Lesser than or equal to 
a = 20
b = 30
is_lesser_than_or_equal = a <= b
print('is_lesser_than_or_equal :', is_lesser_than_or_equal)

```

```python
# Logical Operators 
# AND , OR , NOT

# AND 
age = 20 
height = 170 

# This and operator is used where all the conditions have to be met otherwise it will return the false 
is_he_eligible = age > 18 and height >= 170
print('and operator usecase: ',is_he_eligible)


# OR this will be used in a place where any one of the condition has to be met, otherwise return false. 
is_he_over_18 = (age > 18) or (height > 180) 
print('or operator usecase: ', is_he_over_18)


# NOT operator -> to inverse the logic we use this operator 
isnt_he_over_18 = not(age>18)
print("isnt_he_over_18 or not operator :" , isnt_he_over_18)
```
