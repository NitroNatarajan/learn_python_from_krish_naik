#### Use cases of List 

```python
# Example 1: Manage a TODO list
to_do_list = ["Buy Groceries" , 'Clean the house', 'Prepare the breakfast', 'Taking shower', 'Schedule a meeting']
# Adding a task 
to_do_list.append('Reporting the status to manager')
to_do_list.append('Pay the bills')

# Remove a completed task
to_do_list.remove('Clean the house')

# Check and remind some of the task 
if 'Pay the bills' in to_do_list:
  print("Don't forget to pay the utility bill")

print('Remaining todo list to do...')
for task in to_do_list:
  print(f'-{task}')
```

```python
# student marks management 
marks = [87,68,78, 97,50,58,99,88,56]

marks.append(81)
no_of_students = len(marks)
avg_marks = sum(marks)/no_of_students
print('No_of_students:' , no_of_students)
print('Average: ', round(avg_marks))

max_marks = max(marks)
min_marks = min(marks)

print('Max:' , max_marks)
print('Min:', min_marks)

less_than_60 = 0
for mark in marks:
  if mark<=60:
    less_than_60 += 1
print('How many students scored less than 60 is ',less_than_60)
print(f'Student scored more than 60 is {no_of_students - less_than_60}')
```

#### Managing the inventory 
Here we can have a list as an inventory and we add the item, remove an item when no stock.
 

```python
# Managing the inventory 
inventory = ['apple','banana','orange','mango','pineapple']

# Add elements to the inventory
inventory.append('guava')

# Remove the inventory elements
inventory.remove('apple')

# Check if an item stock is available 
item = 'banana'
if item in inventory:
  print(f"Yes..{item}'s In stock")
else:
  print(f'No.{item} is not available at the moment')

print('Printing the inventory list:-')
for item in inventory:
  print(f'- {item}')
```

```python
#### collecting and analyzing the user feedback 
feedbacks = ['Great service', 'Very satisfied', 'Could be better','Excellent experience']

# Adding new feedback
feedbacks.append('Happy with the service')

# Counting specific feedback
good_list = ['good','nice','excellent','great']
positive_count = 0 
for feedback in feedbacks:
  for positive_comment in good_list:
    if positive_comment.lower() in feedback.lower():
      positive_count += 1 

negative_count = len(feedbacks) - positive_count
# Printing the feedback counts 
print('Positive feedbacks: ' , positive_count)
print('Negative feedback: ', negative_count)
```
