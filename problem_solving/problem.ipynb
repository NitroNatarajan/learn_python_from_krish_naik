### solid square pattern

```python
'''
input -> n integer 
output is a list filled with n-strings, each strings has to be n times of the * 
'''
def print_list(n):
    result = []
    for _ in range(n):
        string = ''
        for _ in range(n):
            string += '*'
        result.append(string)
    return result;

user_input = int(input('Enter the integer input:'))
print(print_list(user_input))
# It is n2 time complexity
```

### Hollow square pattern

```python

def hollow_square(n):
  arr = []
  for i in range(n):
    string = ''
    for j in range(n):
      if i==0 or i == n-1:
        string += '*'
      elif j>0 or j<n-1:
        if j==0 or j == n-1:
          string += '*'
        else:
          string += " "
    arr.append(string)
  return arr

x = hollow_square(5)
print(x)
print(len(x[0]))
print(len(x[1]))           
  
```

```python
def generate_rectangle(n, m):
    """
    Function to return a rectangle pattern of '*' with length n and breadth m as a list of strings.
    
    Parameters:
    n (int): The number of rows in the rectangle.
    m (int): The number of columns in the rectangle.
    
    Returns:
    list: A list of strings where each string represents a row of the rectangle pattern.
    """
    arr = ''
    for i in range(n):
        if i==0 or i==n-1:
            arr += "*"*m 
            arr += '\n'
        else:
            string = ''
            for j in range(m):
                if j == 0 or j==m-1:
                    string += '*'
                else:
                    string += " "
            arr += string
            arr += '\n' 
    return arr

print(generate_rectangle(3,7))
```

```python
def generate_rectangle(n, m):
    """
    Function to return a rectangle pattern of '*' with length n and breadth m as a list of strings.
    
    Parameters:
    n (int): The number of rows in the rectangle.
    m (int): The number of columns in the rectangle.
    
    Returns:
    list: A list of strings where each string represents a row of the rectangle pattern.
    """
    arr = []
    for i in range(n):
        if i==0 or i==n-1:
            arr.append('*'*m)
        else:
            string = ''
            for j in range(m):
                if j == 0 or j==m-1:
                    string += '*'
                else:
                    string += " "
            arr.append(string) 
    return arr

print(generate_rectangle(4,5))
```
