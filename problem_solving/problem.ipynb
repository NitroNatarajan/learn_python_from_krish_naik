```python
'''
input -> n integer 
output is a list filled with n-strings, each strings has to be n times of the * 
'''
def print_list(n):
    result = []
    for _ in range(n):
        string = ''
        for _ in range(n):
            string += '*'
        result.append(string)
    return result;

user_input = int(input('Enter the integer input:'))
print(print_list(user_input))
# It is n2 time complexity
```

```python
def reverse_list(lst):
    # Your code goes here
    length = len(lst)
    result = []
    for i in range(length):
        result.append(lst[length-i-1])
    return result

reverse_list([1, 2, 3, 4, 5,6,7,8])
```

```python
def merge_dicts_with_overlapping_keys(dicts):
  # Your code goes here
  output = {}
  for x in dicts:
    for key, value in x.items():
      if key not in output:
        output[key] = value
      else:
        output[key] += value

  return output

print(merge_dicts_with_overlapping_keys([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}]))
```

```python
def int_to_binary(n):
    """
    Function to convert an integer to its binary representation.
    
    Parameters:
    n (int): The integer to convert.
    
    Returns:
    str: The binary representation of the integer.
    """
    # Your code here
    result = ''
    def remainder(num):
        return num%2
       
    while (n/2) < 2 :
        result += str(remainder(n))
        n /= 2
        
    return result[::-1]

print(int_to_binary(10))
```
